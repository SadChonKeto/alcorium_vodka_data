```python
import requests
from bs4 import BeautifulSoup

baseurl = 'https://alcorium-store.ge/'
headers_h = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'
}
productlinks=[]
for i in range(1, 15):
    r = requests.get(f'https://alcorium-store.ge/%E1%83%90%E1%83%A0%E1%83%90%E1%83%A7%E1%83%98/page-{i}/')
    soup = BeautifulSoup(r.content, 'lxml')

    productlist = soup.find_all('div', id='categories_view_pagination_contents')
    for item in productlist:
        for link in item.find_all('a', href=True):
            productlinks.append(link['href'])
print(productlinks)
```

```python
#testlink = ('https://alcorium-store.ge/%E1%83%90%E1%83%A0%E1%83%90%E1%83%A7%E1%83%98/finlandia/finlandia-1-l/')
vodkalist=[]
for link in productlinks:
    r = requests.get(link, headers=headers_h)
    soup=BeautifulSoup(r.content, 'lxml')
    name = soup.find('h1', class_='ut2-pb__title').text.strip() ## .strip აცლის ზედმეტ სფეისებს
    price = soup.find('div', class_='ut2-pb__price-actual').text.strip()
    info=soup.find('div', class_='ty-features-list').text.strip()
        #print(name, price, info, sep='\n')
    vodka={
        'name': name,
        'price': price,
        'info': info
    }
    vodkalist.append(vodka)
```

```python
import pandas as pd
df=pd.DataFrame(vodkalist)
file_path = 'scraped_data.csv'
df.to_csv(file_path, encoding='utf-8', index=False)
```
