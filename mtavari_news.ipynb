```python
import requests
from bs4 import BeautifulSoup

baseurl = 'https://mtavari.tv'
headers_h = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'
}
links=[]
for i in range(1, 10):
    r = requests.get(f'https://mtavari.tv/news/archive?page={i}')
    soup = BeautifulSoup(r.content, 'lxml')
    productlist = soup.find_all('div', class_='archive__Content-fe26lv-5 cjXGyg')
    for item in productlist:
        for link in item.find_all('a', href=True):
            links.append(baseurl+link['href'])
print(links)
```

```python
for link in links:
    r = requests.get(link, headers=headers_h)
    soup=BeautifulSoup(r.content, 'lxml')
    title = soup.find('h1', class_='id__Title-bhuaj0-10 bWQKRB').text.strip() ## .strip აცლის ზედმეტ სფეისებს
    article=soup.find('div', class_='EditorContent__EditorContentWrapper-ygblm0-0 dzgBwY').text.strip()
        #print(name, price, info, sep='\n')
    comp_article={
        'title': title,
        'main_txt': article
    }
    print(comp_article)
```
